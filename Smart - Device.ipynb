```python
class SmartDrive:
    def __init__(self,device_id,power_state,name):
        self.__device_id = device_id
        self.__power_state = power_state
        self.name = name

    @property
    def device_id(self):
        return self.__device_id

    @property
    def power_state(self):
        return self.__power_state

    @device_id.setter
    def device_id(self, device_id):
        if device_id == "":
            print(f"Device id, {device_id}")
        else:
            print("invalid device id")

    @power_state.setter
    def power_state(self, power_state):
        if power_state == "ON":
            print(f"Power state, {power_state}")
        else:
            print("invalid power state")

    def turn_on(self):
        self.__power_state = "ON"

    def turn_off(self):
        self.__power_state = "OFF"

    def display_info(self):
        print(f"Device ID: {self.__device_id}")
        print(f" Power State: {self.__power_state}")
        print(f"Name: {self.name}")


class TemperatureSensor(SmartDrive):
    def __init__(self, device_id, power_state, name, temperature):
        super().__init__(device_id, power_state, name)
        self.temperature = temperature

    def read_temperature(self):
        if self.power_state == "ON":
            print("reading temperature")
        else:
            print("invalid power state")

    def display_info(self):
        super().display_info()
        print(f"Temperature: {self.temperature}")


class SecurityCamera(SmartDrive):
    def __init__(self, device_id, power_state, name, recording_status):
        super().__init__(device_id, power_state, name)
        self.recording_status = recording_status

    def start_recording(self):
        self.recording_status = "Recording"
        print("Recording Started")

    def stop_recording(self):
        self.recording_status = "Stopped"
        print("Recording Stopped")

    def display_info(self):
        super().display_info()
        print(f"Recording Status: {self.recording_status}")

class SmartLight(SmartDrive):
    def __init__(self, device_id, power_state, name, brightness):
        super().__init__(device_id, power_state, name)
        self.brightness = brightness

    def increase_brightness(self):
        if self.brightness > 0:
            print("Brightness increased")

    def decrease_brightness(self):
        if self.brightness == 0:
            print("Brightness decreased")

    def display_info(self):
        super().display_info()
        print(f"Brightness: {self.brightness}")

temp = TemperatureSensor("GH499", "OFF", "Room Temperature Sensor", 25)
temp.display_info()

light = SmartLight("RM885", "OFF", "Light", 0)
light.display_info()

camera = SecurityCamera("DW664", "OFF", "Camera", "Stopped" )
camera.display_info()


#menu-driven interface
sensor = TemperatureSensor("GH449", "ON", "Room Temperature Sensor", 25)
light = SmartLight("RM885", "ON", "Light", 0)
camera = SecurityCamera("DW664", "ON", "Camera", "Stopped")

while True:
    print("\n===== SMART DEVICE Management System =====")
    print("1. Display Device Information")
    print("2. Turn Device On")
    print("3. Turn Device Off")
    print("4. Read Temperature")
    print("5. Adjust Brightness")
    print("6. Start Recording")
    print("7. Exit")

    choice = int(input("Enter your choice(1-7): "))
    if choice == 1:
        print("\nTemperature Sensor:")
        sensor.display_info()
        print("\nSmart Light:")
        light.display_info()
        print("\nSecurity Camera:")
        camera.display_info()
    elif choice == 2:
        device = input("Enter Device(sensor/light/camera): ")
        if device == "sensor":
            sensor.turn_on()
        elif device == "light":
            light.turn_on()
        elif device == "camera":
            camera.turn_on()
        else:
            print("invalid device")
    elif choice == 3:
        device = input("Enter Device(sensor/light/camera): ").lower()
        if device == "sensor":
            sensor.turn_off()
        elif device == "light":
            light.turn_off()
        elif device == "camera":
            camera.turn_off()
        else:
            print("invalid device")
    elif choice == 4:
        sensor.read_temperature()
    elif choice == 5:
        brightness = int(input("Enter Brightness(0-100): "))
        light.brightness = brightness
    elif choice == 6:
        camera.start_recording()
    elif choice == 7:
        print("Exit")
        break
    else:
        print("invalid choice. Please enter a valid choice")
```
